```python
import pandas as pd
peak_df = pd.read_csv('/content/drive/MyDrive/SK하이닉스/04. 피크데이 계산/PeakDays/PeakDays.csv')
peak_df
```

```python
new_columns = ['date', 'Volume', '6months_Bins']
peak_df.columns = new_columns
peak_df
```

```python
FVE_df = peak_df.drop(['Volume','6months_Bins'], axis=1)
FVE_df
```

```python
FVE_df['date'] = pd.to_datetime(FVE_df['date'], format='%Y%m%d')
FVE_df['date'] = FVE_df['date'].dt.date
FVE_df['date'] = FVE_df['date'].astype(str)
FVE_df
```

```python
topic_list = [1,2,3,4,6,7,9,10,11,14,15,16,19,20,21,28,33,35,39,44,47,49]
for i in topic_list:
  topic_df = pd.read_csv(f'/content/drive/MyDrive/SK하이닉스/07. LASSO 회귀/토픽별 일간 예측된 거래량/topic{i}_pred_volume.csv')
  topic_df['date'] = topic_df['date'].astype(str)
  merged_df = pd.merge(FVE_df, topic_df, on='date', how='left')
  new_column_data = []
  new_column_data_ver2 = []
  for k in range(len(merged_df)):
    new_column_data.append(merged_df['pred_trading_vol_scaled'][k]/merged_df['trading_volume_scaled'][k])
  FVE_df[f'토픽{i}'] = new_column_data

FVE_df
```

```python
file_path = '/content/drive/MyDrive/SK하이닉스/08. FVE 계산/피크일-토픽 별 FVE.csv'
FVE_df.to_csv(file_path, index=False)
```
